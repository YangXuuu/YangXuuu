```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

```python
api_url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&outputsize=full&apikey=demo"
```

```python
res = requests.get(api_url)
res
### request return by the server ###
```

```python
#help(res)
```

```python
res_json = res.json()
```

```python
res_json.keys()
```

```python
res_json['Meta Data']
```

```python
df = pd.DataFrame.from_dict(res_json['Time Series (Daily)'])
df
```

```python
df = pd.DataFrame.from_dict(res_json['Time Series (Daily)']).T
df
```

```python
# 3 things to do when working with new data
# describe
# check missing values
# check data type
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['1. open'] = df['1. open'].astype(float)
df['2. high'] = df['2. high'].astype(float)
df['3. low'] = df['3. low'].astype(float)
df['4. close'] = df['4. close'].astype(float)
df['5. volume'] = df['5. volume'].astype(float)
df.dtypes
```

```python
df.describe()
```

```python
plt.figure(figsize = (8,4.5))
df['1. open'].plot()
```

```python
df.index
```

```python
df.index = pd.to_datetime(df.index)
df.index
```

```python
plt.figure(figsize = (8,4.5))
df['1. open'].plot()
```

```python
# we can use resampling to get a summary of a timeframe
# e.g. we can resample the dataframe to get the week's/month's/year's average
# this takes the average of the week
df.resample("W").mean()['1. open'].plot()
#df.resample("2W").mean()
```

```python
api_url2 ="https://api.disneyapi.dev/characters"
disney_res = requests.get(api_url2)
disney_res
```

```python
# declare a base api string
# declare empty list to store dataframe
df_list = []
# loop from page 1 to page 149
base_url = "https://api.disneyapi.dev/characters?pages="
# in each loop
#     make an api call
#.    turn the response into a dataframe
#.    add the dataframe into a list
for page in range(1,150):
  #print(base_url + str(page))
  temp_res = requests.get(base_url + str(page))
  temp_json = temp_res.json()
  temp_df = pd.DataFrame.from_dict(temp_json['data'])
  df_list.append(temp_df)
# concatenate all of the dataframe into a large dataframe
disney_df = pd.concat(df_list,ignore_index = True)
```
